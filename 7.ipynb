```python
import re
import pandas as pd
import tweetnlp
import spacy
import nltk
from nltk.tokenize import word_tokenize
from string import punctuation
```

```python
ner = tweetnlp.load_model('ner')
nlp = spacy.load("en_core_web_sm")
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/BSProject/datasets/topics25_pre.csv")
tweets = df["Tweet"].tolist()
topics = df["Topic"].tolist()
```

```python
tweets = [re.sub(r"(\n\s*)+", "\n", tweet) for tweet in tweets]
tweets = [re.sub(r"\n$", "", tweet) for tweet in tweets]
tweets = [re.sub(r"^\n", "", tweet) for tweet in tweets]
```

```python
tweets = [re.sub(r"[ \t]+", " ", tweet) for tweet in tweets]
tweets = [re.sub(r"^\s", "", tweet) for tweet in tweets]
tweets = [re.sub(r"\s$", "", tweet) for tweet in tweets]
```

```python
def deleter(tweet):
    tweet = re.sub(r"/", " / ", tweet)
    tweet = re.sub(r",", " , ", tweet)
    t = ner.ner(tweet)
    propn_tweet = []
    for i in t:
        if i['type'] == 'person' or i['type'] == 'group':
            e = i['entity'].strip()
            if e not in propn_tweet:
                propn_tweet.append(e)
    if len(propn_tweet) < 4:
        return 1
    tokens = word_tokenize(tweet)
    tags = nltk.pos_tag(tokens, tagset = "universal")
    tags = [text for (text, pos) in tags if nlp(text.lower()).has_vector and len(text)>2 and text.isalnum() and text not in punctuation and pos == "NOUN"]
    counter = 0
    for i in range(len(propn_tweet)):
        if i != len(propn_tweet)-1:
            if counter == 3:
                return 0
            else:
                e1 = propn_tweet[i]
                e2 = propn_tweet[i+1]
                regex = f"{re.escape(e1)}(.+?\n*){re.escape(e2)}"
                mch = re.search(regex, tweet)
                if mch:
                    items = mch.group(1).split()
                    flag = 1
                    for item in items:
                        if item in tags and len(propn_tweet[i+1:]) < 4:
                            return 1
                        elif item in tags:
                            flag = 0
                    if flag == 1:
                        counter += 1
                    elif counter != 0:
                        counter = 0
    return 0
tweets2 = []
preprocessed2 = []
for i in range(len(tweets)):
    if deleter(tweets[i]) == 1:
        tweets2.append(tweets[i])
        preprocessed2.append(preprocessed[i])
tweets = tweets2
preprocessed = preprocessed2
```

```python
for i in range(len(tweets)):
    temp_list = tweets[i].split('\n')
    for item in temp_list:
        if bool(re.match('(\w|\W)+[a-zA-Z]', item)) == False:
            temp_list.remove(item)
    tweets[i] = "\n".join(temp_list)
```

```python
df = pd.DataFrame({"Tweet": tweets, "Preprocessed": preprocessed})
df.to_csv("/content/drive/MyDrive/Colab Notebooks/BSProject/datasets/ldamss_preprocessed.csv", index=False)
```

```python
!pip install transformers
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification

# load BERT model and tokenizer
model_name = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=2)

# define labels for subevent extraction
labels = ['O', 'SUBEVENT']

# tokenize text
text = "Chelsea beat Manchester United in the Premier League."
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

# add padding to tokens
max_len = 32
padding_length = max_len - len(token_ids)
token_ids += [0] * padding_length
attention_mask = [1] * len(token_ids)

# convert to tensors
input_ids = torch.tensor([token_ids])
attention_mask = torch.tensor([attention_mask])

# pass through model and get predictions
with torch.no_grad():
    outputs = model(input_ids, attention_mask=attention_mask)

predictions = torch.argmax(outputs[0], dim=2).tolist()[0]

# extract subevents from predictions
subevents = []
for i, prediction in enumerate(predictions[:len(tokens)]):
    if labels[prediction] == 'SUBEVENT':
        subevent = tokens[i]
        for j in range(i+1, len(predictions)):
            if labels[predictions[j]] == 'SUBEVENT':
                subevent += " " + tokens[j]
            else:
                break
        subevents.append(subevent)

print(subevents)
```
